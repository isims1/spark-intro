```python
print("Using Pyspark transformations and actions")
print("")

# Load the data
fakeData = scSpark.read.csv("Fake_data.csv", header=True, inferSchema = True)
fakeData.show(3)

cntryCnts = fakeData.groupBy('Birth_Country').count()

print("#1) Find birth country which has highest amount of people")

cntryCnts = fakeData.groupBy('Birth_Country').count()
cntryCnts.orderBy('count', ascending=False).show(1)

print("#2) Find average income of people who are born in united states of america")

filtCntry = fakeData.filter(fakeData["Birth_Country"] == "United States of America")

sumInc = filtCntry.groupBy("Birth_Country").sum().select("sum(Income)")

cntUS = filtCntry.groupBy("Birth_Country").count().select("count")

print("")
print(round(sumInc.collect()[0][0] / cntUS.collect()[0][0],0))
print("")

print("#3) How many people has income over 100,000 but their loan is not approved.")

filtLoan = fakeData.filter(fakeData["Loan_Approved"] == False)
filtInc = filtLoan.filter(filtLoan["Income"] > 100000)

print("")
print(filtInc.count())
print("")

print("#4) Find top 10 people with highest income in United States of america. (Print their names, income and jobs)")

filtCols = fakeData.select("First_Name","Last_name", "Income", "Job")
filtUS = filtCols.filter(fakeData["Birth_Country"] == "United States of America")
filtUS.orderBy("Income", ascending=False).show(10)

print("#5) How many number of distinct jobs are there?")

filtJob = fakeData.select("Job").groupBy("Job").count()

print("")
print(filtJob.count())
print("")

print("#6) How many writers earn less than 100,000?")

filtJobInc = fakeData.select("Job", "Income").filter(fakeData["Income"] < 100000)

print("")
print(filtJobInc.filter(filtJobInc["Job"] == "Writer").count())
print("")
```
